```python
import matplotlib.pyplot as plt
from scipy.stats import t
from scipy.stats import norm
import numpy as np
from ipywidgets import *
```

```python
%matplotlib notebook

x = np.linspace(-13, 13, 200)
fig_t   = plt.figure(figsize=(9.62,6))
ax_t    = fig_t.add_subplot(1, 1, 1)
line_t, = ax_t.plot(x, t.pdf(x, 1),
                   'r-', label='Распределение Стьюдента')

plt.plot(x, norm.pdf(x), 'g-', lw=3, alpha=0.4, label='Нормальное распределение')

def update(df = 1.0):
     
    line_t.set_ydata(t.pdf(x, df))
    ax_t.set_ylim(-0.01, max(1.1 * max(t.pdf(x, df)), 1.03 * max(norm.pdf(x))))
    fig_t.canvas.draw_idle()

    
plt.legend(fontsize = 8)
interact(update, df=widgets.FloatSlider(min=0.001, max=30.0, step=0.5, value=1.0))
```
