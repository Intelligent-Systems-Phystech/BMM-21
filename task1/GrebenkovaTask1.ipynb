```python
%matplotlib notebook
from ipywidgets import *
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as st
import seaborn as sns
import pandas as pd
```

```python
alpha = 0.6
x = np.linspace(-10, 10, 2000)
y1 =[alpha * st.norm.pdf(v, loc=0, scale=1) for v in x]
y2 = [(1 - alpha) * st.norm.pdf(v, loc=5, scale=2) for v in x]
y = np.sum([y1, y2], axis=0)

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
line, = ax.plot(x, y, linewidth=4)
line1, = ax.plot(x, y1)
line2, = ax.plot(x, y2)

def update(alpha = 0.5):
    y1 =[alpha * st.norm.pdf(v, loc=-2, scale=1) for v in x]
    y2 = [(1 - alpha) * st.norm.pdf(v, loc=2, scale=2) for v in x]
    y = np.sum([y1, y2], axis=0)
    line.set_ydata(y)
    line1.set_ydata(y1)
    line2.set_ydata(y2)
    fig.canvas.draw_idle()

interact(update, df=widgets.FloatSlider(min=0.0, max=1.0, step=0.05, value=1.0));
```
